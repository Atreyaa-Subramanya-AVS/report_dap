2. Positive Correlations between Performance and Price:

This implies that vehicles with higher top speeds, longer ranges, and faster charging capabilities tend to be more expensive. These features are likely valued by consumers and contribute to premium pricing.

3. Efficiency Shows Moderate Correlations:

-> Efficiency_WhKm has moderate positive correlations with TopSpeed_KmH and Range_Km (0.5250 and 0.4342, respectively), suggesting some association between energy efficiency and these performance metrics. However, its correlation with PriceEuro is also moderate (0.6876), indicating that efficiency may not be the sole driver of cost.

4. Seating Capacity Has Limited Correlations:

-> Seats has relatively weak correlations with most other features, except for a moderate positive correlation with Range_Km (0.3245). This suggests that seating capacity may not significantly influence performance or price in this dataset. It could be a more independent factor based on vehicle size and intended use.

## Regression Trend

```python
# Calculate the mean of AccelSec and TopSpeed_KmH
mean_AccelSec = df['AccelSec'].mean()
mean_TopSpeed_KmH = df['TopSpeed_KmH'].mean()

# Calculate the slope (m) and y-intercept (b) of the regression line
m = ((df['AccelSec'] - mean_AccelSec) * (df['TopSpeed_KmH'] - mean_TopSpeed_KmH)).sum() / ((df['AccelSec'] - mean_AccelSec) ** 2).sum()
b = mean_TopSpeed_KmH - m * mean_AccelSec

# Predict TopSpeed_KmH values using the regression line
df['TopSpeed_KmH_pred'] = m * df['AccelSec'] + b

# Plot the data points and regression line
plt.scatter(df['AccelSec'], df['TopSpeed_KmH'], color='black', label='Data Points')
plt.plot(df['AccelSec'], df['TopSpeed_KmH_pred'], color='blue', label='Regression Line', linewidth=3)
plt.xlabel('AccelSec')
plt.ylabel('TopSpeed_KmH')
plt.title('Linear Regression for AccelSec vs TopSpeed_KmH')
plt.legend()
plt.show()
# plt.figure(figsize=(3,4))
```

This regression graph visually confirms the strong negative correlation between acceleration and top speed observed in the correlation matrix. It also highlights that while acceleration is a significant factor influencing top speed, other factors likely play a role as well.

```python
# Calculate the mean of AccelSec and Efficiency_WhKm
mean_AccelSec = df['Range_Km'].mean()
mean_TopSpeed_KmH = df['Efficiency_WhKm'].mean()

# Calculate the slope (m) and y-intercept (b) of the regression line
m = ((df['Range_Km'] - mean_AccelSec) * (df['Efficiency_WhKm'] - mean_TopSpeed_KmH)).sum() / ((df['Range_Km'] - mean_AccelSec) ** 2).sum()
b = mean_TopSpeed_KmH - m * mean_AccelSec

# Predict TopSpeed_KmH values using the regression line
df['Efficiency_WhKm_pred'] = m * df['Range_Km'] + b

# Plot the data points and regression line
plt.scatter(df['Range_Km'], df['Efficiency_WhKm'], color='black', label='Data Points')
plt.plot(df['Range_Km'], df['Efficiency_WhKm_pred'], color='blue', label='Regression Line', linewidth=3)
plt.xlabel('Range_Km')
plt.ylabel('TopSpeed_KmH')
plt.title('Linear Regression for Range_Km vs Efficiency_WhKm')
plt.legend()
plt.show()
```

The regression line exhibits a negative slope, indicating a negative relationship between range (Range_Km) and top speed (TopSpeed_KmH). This means that as the range of a car increases, its predicted top speed tends to decrease.

## Covariance 

```python
df.cov()
```

1. Acceleration and Top Speed Exhibit Negative Covariance: 

-> The covariance between AccelSec and TopSpeed_KmH is -103.37, indicating a negative relationship. This suggests that as acceleration time decreases (faster acceleration), top speed tends to decrease as well. This aligns with the trade-off we observed in both the correlation matrix and regression graph.

2. Performance Features Share Positive Covariances: 

-> TopSpeed_KmH, Range_Km, and FastCharge_KmH all have positive covariances with each other, ranging from 4,099.80 to 6,977.99. This implies that these performance features tend to vary together—cars with higher top speeds often have longer ranges and faster charging capabilities.

3. Efficiency and Some Performance Features Covary: 

-> Efficiency_WhKm has positive covariances with TopSpeed_KmH (458.22) and Range_Km (1,166.48), suggesting a tendency for more efficient cars to have higher top speeds and longer ranges. This aligns with expectations in electric vehicle design.

4. Seating Capacity Has Limited Covariances: 

-> Seats has relatively weak covariances with most other features, except for a moderate positive covariance with Range_Km (30.10). This reinforces the idea that seating capacity is a more independent factor in this dataset, not strongly linked to performance or price.
